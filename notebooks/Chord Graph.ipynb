```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
# visualise library
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from networkx.drawing.nx_agraph import graphviz_layout
import seaborn as sns
```

```python
example = pd.read_hdf("/home/nikoleta/Desktop/binomial_after.h5")
example.columns
```

```python
G = nx.Graph()
```

```python
nodes = example.player_name
```

```python
G.add_nodes_from(nodes)
```

```python
edges = []
for i in range(len(example)):
    x, y, z, j = example.neighbors[i][1:][:-1].split(",")
    edges.append(tuple((x, example.player_name[i])))
    edges.append(tuple((y, example.player_name[i])))    
    edges.append(tuple((z, example.player_name[i])))
    edges.append(tuple((j, example.player_name[i])))
    
    
```

```python
G.add_edges_from(edges)
```

```python
colors = np.random.rand(1)
```

```python
plt.figure(figsize=(30,30))
pos = nx.circular_layout(G)
nx.draw_networkx_nodes(G,pos,node_size=100,nodelist=G.nodes(), node_color= 'b', node_shape='v')
nx.draw_networkx_edges(G,pos,alpha=0.09)
plt.show()
```
