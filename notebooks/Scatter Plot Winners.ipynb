```python
import pandas as pd
import numpy as np
```

```python
# visualise libraries
import matplotlib.pyplot as plt
%matplotlib inline
from networkx.drawing.nx_agraph import graphviz_layout
import seaborn as sns
```

```python
data = pd.read_csv("/home/nikoleta/src/jobs/data/ch3_experiment/after_analysis/ratio-rr-fifty.csv")
title = 'Round Robin'
size = '50'
```

```python
# rank data 
data = data.sort('ratio')
```

```python
nplayers = len(data['player_name'].unique())
width = max(nplayers / 3, 12)
height = width / 2
spacing = 4
positions = spacing * np.arange(1, nplayers + 1, 1)
```

```python
# point plot 
figure = plt.figure(figsize=(width, height))
plt.xticks(positions, rotation=90)
plt.xlim(0, spacing * (nplayers + 1))
plt.tick_params(axis='both', which='both', labelsize=8)
plt.xlabel('Strategies', fontsize=20)
sns.pointplot(data= data, x = 'player_name', y='ratio', palette="Blues_d")
plt.title('Winning Ratio for {} Topology Size {}'.format(title,size), fontsize= 50)
plt.savefig("/home/nikoleta/src/Dissertation/src/img/appendix/winners-{}-{}.pdf".format(title,size), format='pdf')
plt.show()
```
