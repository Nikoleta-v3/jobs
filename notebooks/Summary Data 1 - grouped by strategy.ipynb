Summary Data 1 - grouped by strategy
====================================
In this notebook, the data generated from the second generation experiment, 
are summarized in one table. The data are being groubed by strategy and measures such as,
median ranking, wins, participation, mean score, average cooperating ratio etc are computed. 
Thus, the table is going to illustrate the overall perfomance of each strategy.

This summary table will be used for statistical analysis. 

```python
import pandas as pd
```

```python
# import after data
binomial = pd.read_hdf("/home/nikoleta/Desktop/binomial_after.h5")
```

```python
binomial.columns
```

```python
median_ranking = binomial.groupby('player_name')['ranking'].median()
win_count = binomial.groupby('player_name')['wins'].max()
participation_count = binomial.groupby('player_name')['frequency'].max()
win_ratio = binomial.groupby('player_name')['ratio'].max()
mean_score = binomial.groupby('player_name')['normalized_average_score'].mean()
cooperation_rating = binomial.groupby('player_name')['cooperating_ratio'].mean()
neighbourhood_score = binomial.groupby('player_name')['average_neighborhood_score'].mean()
```

```python
df = pd.concat([median_ranking, win_count, participation_count, 
                win_ratio, mean_score, cooperation_rating, neighbourhood_score], axis=1)
```

```python
# classification based on cooperating ratio
classification = []
for i in range(len(cooperation_rating)):
    if cooperation_rating[i] <= 0.2 :
        classification.append(0) #low
    if cooperation_rating[i] > 0.2 and cooperation_rating[i] <= 0.4 :
         classification.append(1) #weak
    if cooperation_rating[i] > 0.4 and cooperation_rating[i] <= 0.6 :
         classification.append(2) #mid
    if cooperation_rating[i] > 0.6 and cooperation_rating[i] <= 0.8 :
         classification.append(3) #moderate
    if cooperation_rating[i] > 0.8 and cooperation_rating[i] <= 1 :
         classification.append(4) #high
df['classification'] = classification
```
